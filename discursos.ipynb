```python
%matplotlib inline
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle
import numpy as np
import pandas as pd
```

```python
discursos = pickle.load(open("discurso.pkl", "rb"))
pontos = pickle.load(open("discurso_points.pkl", "rb"))
pts = np.array(pontos)
dfdiscursos = pd.DataFrame(discursos)
```

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(pts)
km = kmeans.predict(pts)
```

## Professor, tivemos dificuldade em escolher uma visualização por causa das várias possibilidades. Testamos diversas e mandamos para o senhor as 3 que consideramos melhores.

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.scatterplot(pts[:, 5], pts[:, 2], s=40, hue="partido", data=dfdiscursos)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 5], centers[:, 2], c='black', s=100, alpha=0.5)
g.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), ncol=1)
plt.title('Análise de similaridade dos discursos dos deputados federais', fontsize=20)
```
